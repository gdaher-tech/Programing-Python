 ## Em grupos, crie um script Python que implementa uma Unidade L√≥gica Aritm√©tica (ULA) simples. Para essa ULA, escolham 2 opera√ß√µes aritm√©ticas, 2 opera√ß√µes l√≥gicas e 1 opera√ß√£o de deslocamento, para ser implementado. Crie um exemplo que parta de 2 n√∫meros decimais

```python
numero1 = float(input('Digite o primeiro n√∫mero decimal: '))
numero2 = float(input('Digite o segundo n√∫mero decimal: '))

# ----- PRIMEIRA OPERA√á√ÉO ARITM√âTICA -----
print("\n[1¬™ OPERA√á√ÉO ARITM√âTICA OBRIGAT√ìRIA]")
print("Escolha uma opera√ß√£o para realizar entre os n√∫meros:")
print(" 1 - Adi√ß√£o")
print(" 2 - Subtra√ß√£o")
print(" 3 - Divis√£o")
print(" 4 - Multiplica√ß√£o")
opcao1 = input("Digite o n√∫mero da opera√ß√£o desejada: ")

if opcao1 == '1':
    resultado1 = numero1 + numero2
    operacao1 = "Adi√ß√£o"
elif opcao1 == '2':
    resultado1 = numero1 - numero2
    operacao1 = "Subtra√ß√£o"
elif opcao1 == '3':
    if numero2 != 0:
        resultado1 = numero1 / numero2
        operacao1 = "Divis√£o"
    else:
        print("Erro: divis√£o por zero.")
        exit()
elif opcao1 == '4':
    resultado1 = numero1 * numero2
    operacao1 = "Multiplica√ß√£o"
else:
    print("Op√ß√£o inv√°lida. Encerrando programa.")
    exit()

print(f"\n‚úÖ Resultado da 1¬™ opera√ß√£o ({operacao1}): {resultado1:.2f}")

# ----- SEGUNDA OPERA√á√ÉO ARITM√âTICA -----
print("\n[2¬™ OPERA√á√ÉO ARITM√âTICA OBRIGAT√ìRIA]")
print("Escolha outra opera√ß√£o para realizar com os mesmos n√∫meros:")
print(" 1 - Adi√ß√£o")
print(" 2 - Subtra√ß√£o")
print(" 3 - Divis√£o")
print(" 4 - Multiplica√ß√£o")
opcao2 = input("Digite o n√∫mero da nova opera√ß√£o: ")

if opcao2 == '1':
    resultado2 = numero1 + numero2
    operacao2 = "Adi√ß√£o"
elif opcao2 == '2':
    resultado2 = numero1 - numero2
    operacao2 = "Subtra√ß√£o"
elif opcao2 == '3':
    if numero2 != 0:
        resultado2 = numero1 / numero2
        operacao2 = "Divis√£o"
    else:
        print("Erro: divis√£o por zero.")
        exit()
elif opcao2 == '4':
    resultado2 = numero1 * numero2
    operacao2 = "Multiplica√ß√£o"
else:
    print("Op√ß√£o inv√°lida. Encerrando programa.")
    exit()

print(f"\n‚úÖ Resultado da 2¬™ opera√ß√£o ({operacao2}): {resultado2:.2f}")

# ----- OPERA√á√ÉO DE DESLOCAMENTO COM << e OR -----
print("\n[OPERA√á√ÉO FINAL OBRIGAT√ìRIA - DESLOCAMENTO BIT A BIT]")

n1 = int(numero1)
n2 = int(numero2)

shift_n1 = n1 << 1
shift_n2 = n2 << 1
deslocamento = shift_n1 | shift_n2

print(f"""
‚û°Ô∏è  Deslocamento √† esquerda:
    {n1} << 1 = {shift_n1}
    {n2} << 1 = {shift_n2}

‚û°Ô∏è  Opera√ß√£o final com OR:
    {shift_n1} | {shift_n2} = {deslocamento}

üß† Representa√ß√£o em bin√°rio:
    {bin(shift_n1)} | {bin(shift_n2)} = {bin(deslocamento)}
""")
```
